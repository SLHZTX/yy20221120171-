Seaborn


基本背景

    Seaborn 是一个用于在 Python 中制作统计图形的库。它基于 matplotlib 构建，并与 pandas 数据结构紧密集成。Seaborn 帮助你探索和理解数据。它的绘图函数可以操作包含整个数据集的数据框和数组，并在内部执行必要的语义映射和统计汇总，以生成信息丰富的图表。其以数据集为导向的声明式 API 让你可以专注于图表中不同元素的含义，而不是如何绘制它们的细节。在大多数情况下使用seaborn就能做出很具有吸引力的图，而使用matplotlib就能制作具有更多特色的图。应该把Seaborn视为matplotlib的补充，而不是替代物。


发展历史

    Seaborn 由 Michael Waskom 于 2014 年创建，目的是为数据科学家和统计学家提供一个更简洁和直观的可视化工具。它最初是为了补充 Matplotlib 的功能，通过添加默认主题和颜色调色板以及高级的统计图表来简化常见的可视化任务。得益于其简洁的语法和强大的功能,自发布以来, Seaborn 迅速发展，成为数据科学社区中广泛使用的工具。12.0

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline  # 显示图表
```

1.基本设置模块

a.设置图表的大小(该设置会在notbook中一直生效)
    sb.set_context(context=None, font_scale=1, rc=None)
    参数:
        context：设置绘图的上下文，可以选择不同的预设情境, 可选的情境有 'paper'、'notebook'、'talk' 和 'poster'
        font_scale：用于缩放字体大小
        rc: 一个可选的参数，用于覆盖默认的绘图参数

```python
# 示例:
x = [1, 3, 5, 7, 9]
y = [1, 3, 5, 7, 9]
sb.set_context('paper', font_scale=1.5)  # 这里是论文情景,字的大小为1.5
sb.lineplot(x=x,y=y)  # 绘制测试用的直线图
```

b.设置图表的风格(该设置会在notbook中一直生效)
    sb.set_style(style=None, rc=None)
    参数:
        style: 指定图表的样式, 可选样式有'darkgrid'、'whitegrid'、'dark'、'white'、'ticks'
        rc: 一个可选的参数，用于覆盖默认的绘图参数

```python
# 示例:
x = [1, 3, 5, 7, 9]
y = [1, 3, 5, 7, 9]
sb.set_style(style='darkgrid')  # 这里是'darkgrid'风格
sb.lineplot(x=x, y=y)  # 绘制测试用的直线图
```

c.设置绘图调色板(该设置会在notbook中一直生效)
    sb.set_palette(palette, n_colors=None, desat=None, color_codes=False)
    参数:
        palette: 指定调色板的名称或颜色列表,一些预设调色板名称如'deep'、'muted'、'pastel'、'bright'、'dark'等
        n_colors：指定调色板中颜色的数量
        desat：指定颜色的去饱和度
        color_codes：如果为 True，则允许在调色板中使用简写颜色代码

```python
# 示例:
sb.set_palette('deep')
x = np.array([i for i in range(1, 11)])
y1 = np.array([np.exp(i)/2000 for i in range(1, 11)])
y2 = np.array([np.log(i) for i in range(1, 11)])
sb.lineplot(x=x, y=y1)
sb.lineplot(x=x, y=y2)
```

d.设置当前坐标轴的风格
    sb.axes_style(style=None, rc=None)
    参数:
        style：指定图表的样式, 可选样式有'darkgrid'、'whitegrid'、'dark'、'white'、'ticks'
        rc: 一个可选的参数，用于覆盖默认的轴参数

```python
# 示例:
x = np.array([i for i in range(1, 11)])
y = np.array([np.exp(i)/3000 for i in range(1, 11)])
with sb.axes_style("white"):  # 当前坐标轴是'white'风格
    sb.lineplot(x=x, y=y)
```

e.统一设置
    sb.set(context='notebook',style='darkgrid',palette='deep',font='sans-serif',font_scale=1,color_codes=True,rc=None)
        参数:
            context：设置绘图的上下文，预设情境同上context参数
            style: 控制绘图的样式,可选样式同上style参数
            palette: 指定调色板的名称或颜色列表,可选名称同上palette参数
            font: 指定字体的名称
            font_scale: 字体大小
            color_codes: 如果为 True，则允许在调色板中使用简写颜色代码
            rc：一个可选的参数,用于覆盖默认的 Matplotlib 参数

```python
# 示例:
# 设置
sb.set(
    context='notebook',
    style='darkgrid',
    palette='deep'
)
# 数据
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [10, 20, 25, 30, 35]
})
# 绘图
sb.lineplot(data=data, x='x', y='y')
```

2.分类图模块

a.创建分类条形图
    sb.barplot(x=None,y=None,hue=None,data=None,order=None,hue_order=None,estimator=<function meanat 0x000001C71D259288>,ci=95,...)
        常用参数:
            x: 指定要绘制在 x 轴上的数据
            y: 指定要绘制在 y 轴上的数据
            hue: 分类列名,可以创建多个系列的条形图
            data: 要绘图的数据集
            order: 指定 x 轴上类别的显示顺序
            hue_order: 指定 hue 变量的类别显示顺序
            estimator: 用于计算 y 变量的聚合统计值的函数
            ci: 置信区间的大小
            orient: 绘图的方向，v 表示垂直，h 表示水平
            color: 元素的颜色
            palette: 不同级别 hue 变量的颜色,调色板

```python
# 示例:
# 要绘图的数据
data = pd.DataFrame({
    'col1': ['A', 'B', 'C', 'A', 'B', 'C'],
    'col2': [10, 20, 15, 25, 30, 20],
    'col3': ['X', 'X', 'X', 'Y', 'Y', 'Y']
})

# 设置条形图
"""
这里的配置是按col3分类画出x与对应y的图像,选择的类别的调色板为'muted', 水平绘制
"""
sb.barplot(x='col1', y='col2', hue='col3', data=data, orient='h', palette='muted')

# 绘图
plt.show()
```

b.创建分类箱形图
    sb.boxplot(x=None, y=None,hue=None,data=None)
        常用参数:
            x: 指定要绘制在 x 轴上的数据
            y: 指定要绘制在 y 轴上的数据
            hue: 分类列名,可以创建多个系列的条形图
            data: 要绘图的数据集
            orient: 绘图的方向，v 表示垂直，h 表示水平
            color: 元素的颜色
            palette: 不同级别 hue 变量的颜色,调色板

```python
# 示例:
# 要绘图的数据
data = pd.DataFrame({
    'col1': [i for i in range(1, 11)],
    'col2': [np.exp(i) for i in range(1, 11)],
    'col3': ['X', 'X', 'X', 'Y', 'Y', 'Y', 'Z', 'Z', 'Z', 'Z']
})

# 设置箱型图
"""
这里的配置是按col3分类画出x与对应y的图像,选择的类别的调色板为'muted', 垂直绘制, 元素为红色
"""
sb.boxplot(x='col1', y='col2', hue='col3', data=data, orient='v', palette='muted', color='r')

# 绘图
plt.show()
```

c.创建分类散点图
    sb.swarmplot(x=None,y=None,hue=None,data=None,...)

            x: 指定要绘制在 x 轴上的数据
            y: 指定要绘制在 y 轴上的数据
            hue: 分类列名,可以创建多个系列的条形图
            data: 要绘图的数据集
            orient: 绘图的方向，v 表示垂直，h 表示水平

            palette: 不同级别 hue 变量的颜色,调色板

```python
# 示例:
# 要绘图的数据
data = pd.DataFrame({
    'col1': [i for i in range(1, 11)],
    'col2': [np.random.randn() for i in range(1, 11)],
    'col3': ['X', 'X', 'X', 'Y', 'Y', 'Y', 'Z', 'Z', 'Z', 'Z']
})

# 设置散点图
"""
这里的配置是按col3分类画出x与对应y的图像,选择的类别的调色板为'dark', 水平绘制, 元素为蓝色
"""
sb.swarmplot(x='col1', y='col2', hue='col3', data=data, orient='h', palette='dark', color='b')

# 绘图
plt.show()
```

3.分布图模块

a.创建直方图
    sb.histplot(data,x=None,y=None,hue=None,kde=None,...)
        常用参数:
            data: 要绘制直方图的数据
            x: 指定要绘制在 x 轴上的变量(选且只能选x)
            Y: 指定要绘制在 y 轴上的变量(选且只能选y)
            hue: 分类列名
            kde:指定是否绘制核密度估计曲线
            palette: 调色板

```python
# 示例:
# 要绘制的数据
data = pd.DataFrame({
    'col1': [np.random.randn() for i in range(1, 11)],
    'col2': ['X', 'X', 'X', 'Y', 'Y', 'Y', 'Z', 'Z', 'Z', 'Z']
})

# 设置直方图
"""
这里的配置是画出x的直方图,选择的类别的调色板为'dark'
"""
sb.histplot(data, x='col1', kde=True, palette='dark')

# 绘图
plt.show()
```

b.创建KDE图
    sb.kdeplot(data,data2=None,shade=False,vertical=False,kernel='gau',...)
        常用参数:
            data, data2: 要绘制直方图的数据
            {  # 注:如果只选其一则是单核密度估计图,二者都选则是双核密度估计图
            x: 指定要绘制在 x 轴上的变量
            Y: 指定要绘制在 y 轴上的变量
            }
            shade: 是否显示阴影
            vertical: 方向，True 表示垂直，False 表示水平
            kernel: 核函数，可选值包括 ‘gau’, ‘cos’, ‘biw’, ‘epa’, ‘tri’, ‘triw’
            bw: KDE 的带宽（bandwidth），可选值包括 ‘scott’, ‘silverman’, scalar, pair of scalars
            palette: 调色板

```python
# 示例:
# 要绘制的数据
data = pd.DataFrame({
    'col1': [np.random.randn() for i in range(1, 101)],
    'col2': [np.random.randn() for i in range(1, 101)]
})

# 设置KDE图
"""
这里的配置是画出x与y的KDE图,选择的类别的调色板为'dark',显示阴影
"""
sb.kdeplot(data, x='col1', y='col2', shade=True, palette='dark')

# 绘图
plt.show()
```

4.关系图模块

a.创建散点图
    sb.scatterplot(x=None,y=None,hue=None,style=None,size=None,data=None,palette=None)
        常用参数:
            data: 要绘制直方图的数据
            x: 指定要绘制在 x 轴上的变量
            Y: 指定要绘制在 y 轴上的变量
            hue: 按列分类
            style: 风格
            size: 点的大小
            palette: 调色盘

```python
# 示例:
# 要绘制的数据
data = pd.DataFrame({
    'col1': [np.random.randn() for i in range(1, 101)],
    'col2': [np.random.randn() for i in range(1, 101)],
    'col3': [np.random.randint(1, 3)  for i in range(1, 101)]
})

# 设置散点图
"""
这里的配置是画出x与y的散点图,选择的类别的调色板为'pastle'
"""
sb.scatterplot(data, x='col1', y='col2', hue='col3', palette='pastel')

# 绘图
plt.show()
```

测试用例1:
